```python
print("Analiza danych projektu MEDICA+")

# przykładowe dane
pacjenci = [
    {"imie": "Jan", "wizyty": 3},
    {"imie": "Anna", "wizyty": 5},
    {"imie": "Piotr", "wizyty": 2},
    {"imie": "Kasia", "wizyty": 4}
]

# liczba pacjentów
liczba_pacjentow = len(pacjenci)

print("Liczba pacjentów:", liczba_pacjentow)

# suma wizyt
suma_wizyt = 0

for pacjent in pacjenci:
    suma_wizyt += pacjent["wizyty"]

print("Łączna liczba wizyt:", suma_wizyt)

# średnia liczba wizyt
srednia = suma_wizyt / liczba_pacjentow

print("Średnia liczba wizyt na pacjenta:", round(srednia, 2))

# pacjent z największą liczbą wizyt
max_wizyty = max(pacjenci, key=lambda x: x["wizyty"])

print("Najwięcej wizyt miał pacjent:", max_wizyty["imie"])
# Analiza danych w Google Colab

# W ramach projektu MEDICA+ wykorzystano środowisko Google Colab do wykonania przykładowej analizy danych pacjentów.

# Notebook zawierał:

# przykładowe dane pacjentów,
# analizę liczby wizyt,
# obliczenie średniej liczby wizyt,
# wyszukiwanie pacjenta z największą liczbą wizyt.

# Celem ćwiczenia było pokazanie możliwości integracji Google Colab z repozytorium GitHub oraz wykorzystania narzędzi analitycznych w projekcie informatycznym.
```
